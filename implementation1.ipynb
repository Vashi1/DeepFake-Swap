# Step 1: Taking the Image Input

```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#Source Image
im_source = cv2.imread('source.jpg')
gr_im_source = cv2.cvtColor(im_source, cv2.COLOR_BGR2GRAY) 
gr_im_source = np.array(gr_im_source)
gr_im_source.reshape((1000, 1000))

#Destination Image
im_dest = cv2.imread('destination.jpg')
gr_im_dest = cv2.cvtColor(im_dest, cv2.COLOR_BGR2GRAY)
gr_im_dest = np.array(gr_im_dest)
```

# Making the Laplacian Reduce Pyramid for the Images


```python
import cv2
import numpy as np

def generate_laplacian_pyramid(image, levels):
    pyramid = [image]
    for i in range(levels):
        image = cv2.GaussianBlur(image, (5, 5), 0)
        image = cv2.resize(image, (image.shape[1] // 2, image.shape[0] // 2))
        pyramid.append(image)
    return pyramid
```

```python
gr_im_source_l = generate_laplacian_pyramid(gr_im_source, 3)
gr_im_dest_l = generate_laplacian_pyramid(gr_im_dest, 3)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax1.imshow(gr_im_source_l[-1])
plt.title("Source")
ax2 = fig.add_subplot(1, 2, 2)
ax2.imshow(gr_im_dest_l[-1])
# plt.imsave(gr_im_dest_l[-1], "Dest_mod.jpg")
plt.title("Destination")
plt.gray()
plt.show()
```
